### From categorical plot Branch B is the branch with lowest rating

#### The plot above shows:
* There are no outliers in the quantity of any product line.
* From 'Sport and Travel' product both gender purchase are the same.
* 'Home and lifestyle' and 'Food and beverages' products were mostly purchased by the Female gender and also in higher        quantity, though the two product line are both useful to both gender but they are mostly used by the Female gender. 
* More than average of Male gender purchase 'Fashion accessories' in higher quantity.

```python
sns.catplot(x="Product line", y="Total", hue="Gender", hue_order=["Male", "Female"], palette="prism", aspect=4, kind="box", data=s_real).set(title="Total and Product line Plot.")
```

```python
sns.catplot(x="Product line", y="Unit price", kind="point", aspect=3.5, data=s_real).set(title="Product line Per Unit Price")
```

```python
sns.catplot(x="Product line", y="Quantity", kind="point", aspect=3.5, data=s_real).set(title="Product line Per Quantity")
```

* There are products in 'Fashion accessories' which unit price are high. As seen in the product line per quantity plot, the   quantity purchase is low, yet unit price is high, this point to the fact that some individual product have high price.
* Fashion accessories Product that was sold should be restock in higher quantity than before, so as to drive in higher       sales than the previous record.
* Electronic accessories products sold are of low prices, as seen in the plot, high quantity sold, but unit price is doing   opposite of the quantity sold.
* Electronic product should be restocked in higher quantity since they are in high demand, more sales can come in from the   product line.

#### Since the highest Total sales record is from Branch C, which is Port Harcourt. From this branch, patterns of products in high demand can be applied to the remaining two branches.

```python
# dataset unique to branch C called "phc"
phc = s_real[s_real["Branch"] == "C"]
```

```python
g = sns.FacetGrid(phc, col="Product line", col_wrap=3, aspect=1.5)
g.map_dataframe(sns.lineplot, x="Quantity",y="Total")
g.set_titles(col_template="{col_name}")
```

* "Food and beverages" products are selling well in Branch C(City=Port Harcourt), supply of such products to the other       branches can be increased, since Humans cannot do without food, by this customers in Lagos and Abuja can buy more           quantity of Food and beverages products, also introducing discount offer because of more supplies will attract more         customers.
* Same approach can be adopted for "Health and beauty" products in Branch A and Branch B.
